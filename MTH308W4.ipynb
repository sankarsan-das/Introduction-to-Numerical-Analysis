```python
#Newton Method 
from tabulate import tabulate
def f(x,a=1,b=-2,c=5,d=-6):
    return a*x**3+b*x**2+c*x+d
def df(x):
    return (f(x+1e-10)-f(x))/1e-10
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.linspace(-3,5,100)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
plt.plot(x,list(map(f,x)))
x_initial=5
x_next=0
int_list=[]
next_list=[]
error_list=[]
for i in range(8):
    x_next=x_initial-f(x_initial)/df(x_initial)
    int_list.append(x_initial)
    next_list.append(x_next)
    error_list.append(abs((x_initial-x_next)/x_next)*100)
    x_initial=x_next
d={"Iteration":[i for i in range(0,len(int_list))],"Initial point":int_list,"Next point":next_list,"Function value":list(map(f,int_list)),"Errors":error_list}
print(tabulate(d,headers="keys",tablefmt="psql"))
    
    
    
    
```

```python
#Secant Method 
from tabulate import tabulate
def f(x,a=1,b=-2,c=5,d=-6):
    return a*x**3+b*x**2+c*x+d
def df(x1,x2):
    return (f(x1)-f(x2))/(x1-x2)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.linspace(-3,5,100)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
plt.plot(x,list(map(f,x)))
x_prev=3
x_initial=5
x_next=0
int_list=[]
prev_list=[]
next_list=[]
error_list=[]
for i in range(8):
    x_next=x_initial-f(x_initial)/df(x_initial,x_prev)
    int_list.append(x_initial)
    next_list.append(x_next)
    prev_list.append(x_prev)
    error_list.append(abs((x_initial-x_next)/x_next)*100)
    x_prev,x_initial=x_initial,x_next
d={"Iteration":[i for i in range(0,len(int_list))],"Prev func":prev_list,"Initial point":int_list,"Next point":next_list,"Function value":list(map(f,int_list)),"Errors":error_list}
print(tabulate(d,headers="keys",tablefmt="psql"))
```
