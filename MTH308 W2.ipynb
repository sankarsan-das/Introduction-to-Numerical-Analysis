```python
import numpy as np
import pandas as pd
def func(x,a=1,b=-0.65,c=0,d=3.567e-4):
    return a*x**3+b*x**2+c*x+d
from tabulate import tabulate
x_l=-10
x_u=12
lower_list=[x_l]
upper_list=[x_u]
middle_list=[(x_l+x_u)/2]
iterations=50
if(func(x_l)*func(x_u)>0):
    print("No root between x_u={} and x_l={} can be found using bisection method".format(x_u,x_l))
else:
    for i in range(iterations):
        x_m=(x_l+x_u)/2
        if(func(x_u)*func(x_m)<0):
            x_l=x_m
            lower_list.append(x_l)
            upper_list.append(x_u)
            middle_list.append((x_u+x_l)/2)
            continue
        elif(func(x_l)*func(x_m)<0):
            x_u=x_m
            lower_list.append(x_l)
            upper_list.append(x_u)
            middle_list.append((x_u+x_l)/2)
            continue
        else:
            print("Root found at x_m={}".format(x_m))
        
d={"Upper value":upper_list,"Lower value":lower_list,"Middle value":middle_list, "f(x_m)":list(map(func,middle_list))}
data=pd.DataFrame(d)
print(data)
print(tabulate(d,headers="keys",tablefmt="sql",showindex="always"))
```
