```python
#Modified Regula Falsi Method 
def f(x,a=1,b=0,c=-5,d=8):
    return a*x**3+b*x**2+c*x+d
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.linspace(-5,5,10000)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
plt.plot(x,list(map(f,x)))
x_low=-4
x_high=-2
x_mrf=0
F=f(x_low)
G=f(x_high)
low_list=[]
high_list=[]
mrf_list=[]
for i in range(10):
    x_mrf=(abs(F)*x_high+abs(G)*x_low)/(abs(F)+abs(G))
    low_list.append(x_low)
    high_list.append(x_high)
    mrf_list.append(x_mrf)
    if(f(x_mrf)*f(x_low)<0):
        x_high=x_mrf
        G=f(x_high)
        F=F/2
    elif(f(x_mrf)*f(x_high)<0):
        x_low=x_mrf
        F=f(x_low)
        G=G/2

print(tabulate({"Lower Point":low_list,"Higher Point":high_list,"Regula Falsi":mrf_list,"Func value":list(map(f,mrf_list))},headers="keys",tablefmt="psql"))
    
    
```

```python
#Newton Raphson Method 
def f(x,a=1,b=0,c=-5,d=8):
    return a*x**3+b*x**2+c*x+d
def df(x):
    return (f(x+1e-10)-f(x))/1e-10
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.linspace(-5,5,10000)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
plt.plot(x,list(map(f,x)))
x=0
x_list=[x]
for i in range(70):
    x=x-f(x)/df(x)
    x_list.append(x)
print(tabulate({"x values":x_list,"function values":list(map(f,x_list))},headers="keys",tablefmt="psql"))
```

```python
# Secant Method
def f(x, a=1, b=0, c=-5, d=8):
    return a * x**3 + b * x**2 + c * x + d

from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np

# Plotting the function
x_vals = np.linspace(-5, 5, 10000)
plt.figure(figsize=(8, 6))
ax = plt.gca()
ax.spines["top"].set_color("none")
ax.spines["left"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["bottom"].set_position("zero")
plt.plot(x_vals, list(map(f, x_vals)), label="f(x)")
plt.title("Secant Method: Finding Roots")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(alpha=0.3)
plt.axhline(0, color="gray", linewidth=0.8, linestyle="--")

# Secant Method Iterations
x = 4
x_prev = -2
x_list = [x]
tolerance = 1e-6
max_iterations = 20

for i in range(max_iterations):
    if abs(f(x)) < 1e-10 or abs(x - x_prev) < tolerance:
        break
    next_x = x - f(x) * (x - x_prev) / (f(x) - f(x_prev))
    x_prev, x = x, next_x
    x_list.append(x)

# Tabulate results
results = {"x values": x_list, "f(x)": [f(val) for val in x_list]}
print(tabulate(results, headers="keys", tablefmt="psql"))

```
